```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
myImage = np.zeros((240,320))
width_frequency = 3.14159/240
frequency = width_frequency * 5

for i in range(240):
    for j in range(320):
        t = math.sin(math.sqrt(i * i + j) * frequency)
        myImage[i,j]=t
            
plt.imshow(myImage, clim=(0,1),cmap="gray")
plt.imshow(myImage, interpolation="bilinear", clim=(0,1),cmap="gray")
plt.show()
```

```python
myImage = np.zeros((500,500))
PI = 3.14159
width_frequency = 3.14159 / 50
frequency = width_frequency * 1000

for i in range(500):
    for j in range(500):
        t = math.sin(i / frequency) * math.tan(j / frequency) + math.cos(j / frequency) * math.tan(i / frequency);
        myImage[i,j]=t
            
plt.imshow(myImage, clim=(0,1),cmap="gray")
plt.imshow(myImage, interpolation="bilinear", clim=(0,1),cmap="gray")
plt.show()
```

```python
myImage = np.zeros((800,1500))
PI = 3.14159
width_frequency = 3.14159 / 100
test = math.sin(math.cos(PI) * 50)
frequency = width_frequency * 1000

for i in range(800):
    for j in range(1500):
        t = math.sin(i / frequency) + math.sin(j / frequency * test) * math.tan(i * frequency) * 50;
        myImage[i,j]=t
            
plt.imshow(myImage, clim=(0,1),cmap="gray")
plt.imshow(myImage, interpolation="bilinear", clim=(0,1),cmap="gray")
plt.show()
```

```python
myImage = np.zeros((800,1500))
PI = 3.14159
width_frequency = 3.14159 / 100
test = math.sin(math.cos(PI) * 50)
frequency = width_frequency * 1000

for i in range(800):
    for j in range(1500):
        t = math.sin(i / frequency) * math.tan(i * frequency) * math.sqrt(i * i + j * j) * test;
        myImage[i,j]=t
            
plt.imshow(myImage, clim=(0,1),cmap="gray")
plt.imshow(myImage, interpolation="bilinear", clim=(0,1),cmap="gray")
plt.show()
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()

x = np.linspace(40, 50)
ncolors = len(plt.rcParams['axes.prop_cycle'])
shift = np.linspace(0, 50, ncolors, endpoint=True)
for s in shift:
    ax.plot(np.cos(x), np.sin(x * s), 'o-')
    
plt.show()
```
